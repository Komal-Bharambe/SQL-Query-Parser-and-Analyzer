```python
from sqlglot import parse_one, exp
import json

sql_query = """
SELECT c.id, c.name, SUM(o.amount) as total_amount
FROM customers c
JOIN orders o ON c.id = o.customer_id
WHERE c.age > 25
GROUP BY c.id, c.name
ORDER BY total_amount DESC
"""

expression = parse_one(sql_query)

# Extract tables
tables = [table.name for table in expression.find_all(exp.Table)]

# Extract columns
columns = [col.name for col in expression.find_all(exp.Column)]

# Extract group by
group_by = [g.sql() for g in expression.find_all(exp.Group)]

# Extract order by
order_by = [o.sql() for o in expression.find_all(exp.Order)]

# Extract functions/aggregations (like SUM, COUNT)
aggregations = [f.sql() for f in expression.find_all(exp.Func)]

# joins_info = []
# for join in expression.find_all(exp.Join):
#     joins_info.append({
#         "join_type": join.args.get("kind") or "INNER",
#         "table": join.this.sql(),  # joined table
#         "condition": join.args.get("on").sql() if join.args.get("on") else None
#     })

joins = [f.sql() for f in expression.find_all(exp.Join)]


result = {
    "tables": tables,
    "Joins": joins_info,
    "columns": columns,
    "group_by": group_by,
    "order_by": order_by,
    "aggregations": aggregations,
    "total_joins": len(joins_info),
    "Join": joins
}

print(json.dumps(result, indent=4))
```
